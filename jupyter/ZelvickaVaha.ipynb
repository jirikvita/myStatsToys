```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

vaha  = [25, 39, 55, 65, 101, 145, 172, 206, 220]
datum = ['24.12.2022', '28.1.2023', '22.3.2023', '20.4.2023', '19.7.2023', '15.10.2023', '18.11.2023', '15.12.2023', '21.1.2024']
date_objects = [datetime.strptime(date, "%d.%m.%Y") for date in datum]

plt.plot(date_objects, vaha, marker='o')

# Beautify the x-labels (optional)
plt.gcf().autofmt_xdate()

# Labeling the plot
plt.title('Vaha zelvicky [g]')
plt.xlabel('Datum')
plt.ylabel('Vaha')
```
