```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

# Define your user-defined function
def my_function(x, A):
    return (1 - np.cos(A*x)) / (A*x*x)

# Generate x values
x_values = np.linspace(-10, 10, 100)  # You can adjust the range and number of points as needed

# Calculate y values by applying the function to each x value
yvals = []
As = [12, 10, 8, 5, 1]
for A in As:
 yvals.append(my_function(x_values, 2*A))


# Plot the function
for yvs,a in zip(yvals,As):
  plt.plot(x_values, yvs, label='A={:1.1f}'.format(a))
plt.title('Plot of User-Defined Function')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
